```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
newd=pd.read_csv("/Users/tejastammewar/Desktop/pythonPractise/canada-data.csv")
newd
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('pci')
plt.scatter(newd.year,newd.percpi,color='red',marker='*')

```

```python
new_df=newd.drop('percpi',axis=1)
new_df
```

```python
percpi=newd.percpi
percpi
```

```python
reg=linear_model.LinearRegression()
reg.fit(new_df,percpi)
```

```python
reg.predict([[2020]])
```

```python
plt.plot(newd.year,reg.predict(newd[['percpi']]),color='blue')
```
